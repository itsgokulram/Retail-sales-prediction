Only 4 main holidays is available in the data.

By comparing the date and week number we can say the following speculation.

**WEEK 6:** Super Bowl is an unofficial national holiday in the United States, which falls in first or second sunday of february

**WEEK 36:** Labor Day is on the first Monday of September every year in U and it falls in week 36 or 37 of the year

**WEEK 47:** Thanksgiving, which occurs on the fourth Thursday in November, which comes in week 47 or week 48

**WEEK 52** Christmas is observed primarily on December 25, which probably comes in the last week of the year


```python
# replacing the IsHoliday values with the respective holiday occasion

df_feature_store['IsHoliday'] = np.where((df_feature_store['IsHoliday'] == True) & (df_feature_store['Week_Number'] == 6), 'Super Bowl',
                                np.where((df_feature_store['IsHoliday'] == True) & (df_feature_store['Week_Number'] == 36), 'Labor Day',
                                np.where((df_feature_store['IsHoliday'] == True) & (df_feature_store['Week_Number'] == 47), 'Thanksgiving',
                                np.where((df_feature_store['IsHoliday'] == True) & (df_feature_store['Week_Number'] == 52), 'Christmas', 'No Holiday'))))
```

Training the model to predict Unemployment

```python
x = df_cpi_unemp_notnull.drop(['CPI', 'Unemployment'], axis = 1)
y = df_cpi_unemp_notnull['Unemployment']
```

```python
# Splitting the data for train test

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# Function to Compare the metrics for all of the models

models = [('XGB Regressor', XGBRegressor()),
          ('Decision Tree Regressor', DecisionTreeRegressor()),
          ('AdaBoost Regressor' ,AdaBoostRegressor()),
          ('Extra Trees  Regressor', ExtraTreesRegressor()),
          ('Random Forest Regressor', RandomForestRegressor()),
          ('Gradient Boosting Regressor', GradientBoostingRegressor()),
          ('HistGradient Boosting Regressor', HistGradientBoostingRegressor()),
          ('Linear Regression', LinearRegression())
          ]

result = []

def ml_regression():

    for model_name, model in models:

        model.fit(x_train, y_train)
        y_train_pred = model.predict(x_train)
        y_test_pred = model.predict(x_test)

        metrics = {'Model name': model_name,
                   'Train_mse': round(mean_squared_error(y_train, y_train_pred), 4),
                   'Test_mse': round(mean_squared_error(y_test, y_test_pred), 4),
                   'Train_mae': round(mean_absolute_error(y_train, y_train_pred), 4),
                   'Test_mae': round(mean_absolute_error(y_test, y_test_pred), 4),
                   'Train_R2_score': round(r2_score(y_train, y_train_pred), 4),
                   'Test_R2_score': round(r2_score(y_test, y_test_pred), 4)}
        result.append(metrics)

    score_df = pd.DataFrame(result)

    return(score_df)
```

```python
ml_regression()
```
